```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import sys
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

from matplotlib.ticker import MultipleLocator
import matplotlib.ticker as ticker
```

```python
polymer="PhPTZ"
solution="EC_DMC_LiPF6"
percent="10_Percent"
temp="300K"

name=f"{polymer}-{solution}"
print(name)
data_path=f"/Volumes/project/depablo/mleyf3/DMREF/{name}/{percent}_0_SoC"

if polymer == "PVMPT":
    data_path=f"/Volumes/project/depablo/mleyf3/DMREF/PVMPT_Simulations/{name}/{percent}_0_SoC"

if percent == "0_Percent":
    data_path=f"/Volumes/project/depablo/mleyf3/DMREF/{polymer}/{polymer}_n30"

print(data_path)
```

```python
orange = '#FF871E'
blue = '#2178B4'
green = '#2EA12E'
purple = '#9568BD'
red = '#D6292A'
```

```python
#polymers=["PhPTZ", "PVMPT", "PVPPP", "PVBPT", "PTZMA"]
polymers=["PhPTZ", "PVMPT", "PVBPT", "PTZMA"]

colors = {
    "PhPTZ":blue,
    "PVMPT":orange,
    "PTZMA":green,
    "PVPPP":purple,
    "PVBPT":red,
}
percent="20_Percent"
averages = []

fig, ax = plt.subplots(figsize=(9, 6))

ax.axhline(1, c="k", ls="--", lw=2, alpha = 0.8)

for polymer in polymers:
    
    name=f"{polymer}-{solution}"
    print(name)
    data_path=f"/Volumes/project/depablo/mleyf3/DMREF/{name}/{percent}_0_SoC"
    
    if polymer == "PVMPT":
        data_path=f"/Volumes/project/depablo/mleyf3/DMREF/PVMPT_Simulations/{name}/{percent}_0_SoC"
        
    rdf = np.genfromtxt(
        [i for i in open(f'{data_path}/cool/rdf/rdf_all_{temp}.xvg').read().splitlines() if not i.startswith(('#', '@'))]
    )[:]
    # Optional: plot each trajectory
    ax.plot(rdf.T[0], rdf.T[1],
            label=polymer,
            c=colors[polymer],
            alpha=0.8, lw=3)



ax.set_xlabel("r (nm)", fontsize=24)
ax.set_ylabel("g(r)", fontsize=24)
#ax.set_title("Redox Unit - Redox Unit", fontsize=22, pad=10)

ax.set_xlim(-0.05, 2.2)

ax.legend(fontsize=20, 
          #bbox_to_anchor=(1, 0.5),
          frameon=False)

for spine in ax.spines.values():
    spine.set_linewidth(2.5)

ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_minor_locator(MultipleLocator(0.25))

ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.25))

ax.tick_params(axis='both', width=2.0, length=7, labelsize=20)
ax.tick_params(axis='both', which='minor', width=2.0, length=4)

plt.savefig(f'/Users/mleyf/Documents/DMREF/Backbone study/Figures/rdf_unit-unit_300K_{percent}.png', bbox_inches='tight', dpi=300)
print(f"{percent} {polymer} {solution} - {temp}")
print("------------------")
print(f"Saved as rdf_unit-unit_300K_{percent}.png")
```
